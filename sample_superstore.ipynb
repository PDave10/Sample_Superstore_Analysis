### Data Cleaning

```python
df_clean = pd.read_csv("df_exploration.csv")
```

```python
df_clean
```

```python
df_clean.info()
```

```python
df_clean.drop("Unnamed: 0",axis=1,inplace=True)
```

```python
df_clean
```

```python
#checking for invalid data in data set
df_clean.isnull()
```

Adding additional column information

```python
df_clean
```

```python
# Ceating the Revenue Column!
df_clean['Revenue'] = df_clean['Sales'] - df_clean['Profit']
```

```python
# Ceating the Gross Price Column!
df_clean['GrossPrice'] = (df_clean['Revenue'] - df_clean['Profit']) / df_clean['Quantity']
```

```python
df_clean.head(2)
```

```python
# Ceating the Net Price Column!
df_clean['NetPrice'] = df_clean['GrossPrice'] - (df_clean['GrossPrice'] * df_clean['Discount'])
```

```python
df_clean.head(2)
```

```python
df_clean.tail(2)
```

```python
# droping the coloumn, 'GrossPrice'
df_clean = df_clean.drop(columns=['GrossPrice'])
```

```python
df_clean
```

```python
df_clean.to_csv('df_cleaned.csv', index = False)
```

### Data Projection

```python
df_dp = pd.read_csv("df_cleaned.csv")
```

```python
df_dp
```

```python
df_dp.info()
```

```python
# What is the average sales by product category?
gradient = np.linspace(0, 4, 11)
colors = [plt.cm.Spectral(i) for i in gradient]
df_dp.groupby("Category")["Sales"].mean().plot.bar(color=colors)
plt.xlabel("Category",labelpad=15)
plt.ylabel("Sales",labelpad=10)
```

```python
# Which shipping method has the highest profit margin?
gradient = np.linspace(0, 1, 11)
colors = [plt.cm.Spectral(i) for i in gradient]
df_dp.groupby("Ship Mode")["Profit"].sum().plot.bar(color=colors)
plt.xlabel("Ship Mode",labelpad=10)
plt.ylabel("Profit",labelpad=10)
```

```python
# Which customer segment is the most profitable?
colors = plt.cm.rainbow(np.linspace(0, 1, 11))
df_dp.groupby("Segment")["Profit"].mean().plot.bar(color=colors)
plt.xlabel("Segment",labelpad=10)
plt.ylabel("Profit",labelpad=10)
```

```python
# What are the top selling products in the country?
df_dp.groupby("Sub-Category")["Sales"].sum().sort_values(ascending=False).plot.bar()
plt.xlabel("Sub-Category",labelpad=10)
plt.ylabel("Sales",labelpad=10)
```

```python
# What is the impact of discount on sales?
df_dp.groupby("Discount")["Sales"].mean().plot.line(color='purple')
plt.xlabel("Discount",labelpad=10)
plt.ylabel("Sales",labelpad=10)
```
